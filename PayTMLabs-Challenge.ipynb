```python
def load_data(data_path):
    dfs = [delayed(pd.read_csv)(os.path.join(data_path, file)) for file in data_filenames]
    ddf = dd.from_delayed(dfs) # ddf is a dask dataframe
    return ddf
    
```

# Step 2

```python
# Make a copy of original ddf
from copy import copy
ddf_copy = copy(ddf)
```

```python
ddf_copy = ddf_copy.dropna()
ddf_copy.isnull().sum().compute()
```

```python
ddf_copy.compute()
```

```python
import re
def str_to_float(s):
    result = re.findall(r"[-+]?\d*\.\d+|\d+", s)
    return float(result[0])
    
```

## 1. Which country had the hottest average mean temperature over the year?

```python
print(str_to_float('-11.7*'))
```

```python
ddf_copy['MAX'].dtype
```

```python
ddf_copy['MAX'] = ddf_copy['MAX'].astype('string')
ddf_copy['MAX'].dtype
```

```python
# Convert to float
# ddf_copy['MAX_f'] =  ddf_copy['MAX'].apply(lambda x: str_to_float(x), meta=('MAX', 'float'))
ddf_copy['MAX_f'] =  ddf_copy['MAX'].apply(lambda x: str_to_float(x), meta=('MAX', 'float'))
```

```python
ddf_copy.compute()
```

```python
ddf_copy.loc[ddf_copy['MAX_f'] == 9999.9].compute()
```

```python
def filter_missing_val(ddf, col, val):
    return ddf[ddf[col] != val].compute()
```

```python
# df_max = ddf_copy[ddf_copy.MAX_f != 9999.9].compute()
df_max = filter_missing_val(ddf_copy, 'MAX_f', 9999.9)
```

```python
df_max.loc[df_max['MAX_f'] == 9999.9]
```

```python
df_max.shape
```

## 1. Which country had the hottest average mean temperature over the year? - Niger

```python
df_max.groupby('COUNTRY_FULL')['MAX_f'].mean().sort_values(ascending=False)
```
